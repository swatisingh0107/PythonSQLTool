# Report Generator
The python script CreateDatabaseBikeSales was used to create a SQL Server DataBase. The tool used was SQL Server Explorer in Visual Studio where a local db instance was created and data was loaded. The following scripts attempts to connect to the sql server and executes queries

```python
import pyodbc 
import pandas as pd


conn = pyodbc.connect( '''Driver={SQL Server Native Client 11.0};
                        Server=(localdb)\ProjectsV13;
                        Database=OrderManagement;
                        port=1433;
                        Trusted_Connection=yes'''   )


cursor = conn.cursor()
cursor.execute('SELECT * FROM OrderManagement.production.stocks')

for row in cursor:
    print(row)
```

```python
# Count total sales per category
query1 = '''select production.categories.category_name,sum(sales.order_items.quantity) as Total_Sales, sum(sales.order_items.list_price) as TotalRevenue
FROM sales.order_items
INNER JOIN production.products
ON sales.order_items.product_id=production.products.product_id
INNER JOIN production.categories
on production.products.category_id=production.categories.category_id
GROUP BY category_name
ORDER BY TotalRevenue DESC;'''
```

```python
CategorySales=pd.read_sql_query(query1,conn)
CategorySales
```

```python
#Total Revenue till date
query2 = '''select '$' + cast(sum(list_price) as varchar(20)) as TotalRevenue
from OrderManagement.sales.order_items
'''

pd.read_sql_query(query2,conn)
```

```python
#Total Sales in last 3 months
query3='''select sum(list_price) as TotalRevenue, count(sales.orders.order_id) as TotalOrders
from sales.orders
INNER JOIN sales.order_items
on sales.orders.order_id=sales.order_items.order_id
WHERE dateadd(day,-90,(select MAX(order_date) from sales.orders))<=sales.orders.order_date;
'''
pd.read_sql_query(query3,conn)

```
